```python
# Imports
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_columns',50)
```

```python
os.makedirs('Data/', exist_ok=True)
os.listdir("Data/")
```

```python
# reading in the data
akas = pd.read_csv('Data/title-akas-us-only.csv', low_memory=False)
basics = pd.read_csv('Data/title.basics.tsv.gz', sep='\t', low_memory=False)
ratings = pd.read_csv('Data/title.ratings.tsv.gz', sep='\t', low_memory=False)
```

```python
# Filter basics down
filter_us_titles = basics['tconst'].isin(akas['titleId'])
basics = basics[filter_us_titles]
basics
```

```python
# Replacing the '\N' strings with actual NaNs
basics = basics.replace({'\\N':np.nan})
basics
```

```python
# Dropping rows with NaN runtimeMinutes
basics = basics.dropna(subset='runtimeMinutes', inplace=False)
```

```python
# Dropping rows with NaN genres
basics = basics.dropna(subset='genres', inplace=False)
basics
```

```python
# Filter to keep only full-length movies
movie_filter = basics['titleType'] =='movie'
basics = basics[movie_filter]
basics
```

```python
# Convert startYear to a float dtype
basics['startYear']= basics['startYear'].astype(float)
basics['startYear'].dtype
```

```python
# Filtering Start Years (inclusive is true by default)
basics = basics[basics['startYear'].between(2000, 2022)]
basics.head()
```

```python
# Removing "Documentary" genre entries
filter_documentaries = basics['genres'].str.contains('Documentary')
basics = basics[~filter_documentaries]
basics.info()
```

```python
# Displaying a Final Preview
basics.head()
```

```python
basics.info()
```

```python
# Saving data to a csv file
fname_out = "Data/basics_filtered.csv"
basics.to_csv(fname_out, index=False)
```

```python
# Prepping Ratings
ratings.head()
```

```python
# Keeping only the movies included in 'basics'
basic_filter = ratings['tconst'].isin(basics['tconst'])
ratings = ratings[basic_filter]
ratings
```

```python
# Replacing the '\N' strings with actual NaNs
ratings = ratings.replace({'\\N':np.nan})
ratings
```

```python
# Saving to a csv file
fname_out = "Data/ratings_filtered.csv"
ratings.to_csv(fname_out, index=False)
```
